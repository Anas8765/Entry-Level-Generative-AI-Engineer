##AI Summarizer

```python
!pip install transformers sentencepiece
```

```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
import torch
```

```python
sentences = []
current_sentence = ""
full_stops_count = 0

for char in text:
    current_sentence += char
    if char == '.':
        full_stops_count += 1
        if full_stops_count == 20:
            sentences.append(current_sentence.strip())
            current_sentence = ""
            full_stops_count = 0

# Add the remaining text as the last sentence if any
if current_sentence:
    sentences.append(current_sentence.strip())
sentences
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/bart-large-cnn"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)



def get_response(input_text):
  batch = tokenizer(input_text, truncation=True, padding='longest', return_tensors="pt")
  translated = model.generate(**batch)
  tgt_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
  return tgt_text
```

```python
from tqdm.auto import tqdm

summarize = []

for i in tqdm(sentences):
  a = get_response(i)
  summarize.append(a)
```

```python
summarize2 = [' '.join(x) for x in summarize]
summarize2
```

```python
summarize3 = [' '.join(x for x in summarize2) ]
summarized_text = str(summarize3).strip('[]').strip("'")
summarized_text
```

```python
from fpdf import FPDF

pdf = FPDF()

pdf.add_page()

pdf.set_font("Arial", size = 12)

for x in summarized_text:
	pdf.cell(200, 10, txt = x, ln = 1, align = 'C')

pdf.output("summarized_book.pdf")
```
